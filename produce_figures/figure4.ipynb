```python
import seaborn as sns
import matplotlib
```

```python
order_list= ['spatialign', 'precast', 'graphST', 'scalex', 'harmony', 'combat', 'BBKNN', 'scanorama', 'mnn']
name_list=['spatiAlign', 'PRECAST', 'GraphST', 'SCALEX', 'Harmony', 'Combat', 'BBKNN', 'Scanorama', 'MNN']
```

```python
moran_df=pd.read_csv("./moran.csv")
plt.figure(figsize=(10, 6))
ax=sns.boxplot(y="Moran",x="name",data=moran_df,order=order_list)
x = np.arange(len(order_list))
plt.xticks(x, labels=name_list, rotation=30, fontsize=18, ha="center")
plt.yticks(fontsize=18)
plt.ylim(0, 1)
plt.grid()
plt.gca().set_axisbelow(True)
plt.box(False)
plt.xlabel(None)
matplotlib.rcParams.update({'figure.autolayout': True})
ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)
# ax.set_xlabel("Method")
ax.set_ylabel("Moran's I", fontsize=18)
plt.savefig("./moran1.pdf",dpi=150,bbox_inches="tight")
# plt.savefig("./figure/figure_d.pdf")
```
